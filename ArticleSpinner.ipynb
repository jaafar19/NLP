```python
import numpy as np
import pandas as pd
import textwrap
import nltk
from nltk import word_tokenize
from nltk .tokenize.treebank import TreebankWordDetokenizer
```

```python
text = df[df['labels'] == 'business']['text']
text.head()
```

```python
prob = {}
for doc in text:
  lines = doc.split('\n')
  for line in lines:
    words = word_tokenize(line)
    for i in range(len(words) - 2):
      T_0 = words[i]
      T_1 = words[i + 1]
      T_2 = words[i + 2]
      key = (T_0, T_2)
      if key not in prob:
        prob[key] = {}
      if T_1 not in prob[key]:
        prob[key][T_1] = 1
      else:
        prob[key][T_1] + 1
```

```python
prob
```

```python
for key, value in prob.items():
  total = sum(value.values())
  for k , v in value.items():
    value[k] = v / total
```

```python
def spin_documents(doc):
  lines = doc.split('\n')
  output = []
  for line in lines:
    if line:
      new_line = spin_line(line)
    else:
      new_line = line
    output.append(new_line)
  return '\n'.join(output)
```

```python
def sample_word(d):
  p0 = np.random.random()
  cumulaive = 0
  for t, p in d.items():
    cumulaive += p
    if p0 < cumulaive:
      return t
  assert(False)
```

```python
def spin_line(line):
  words = word_tokenize(line)
  i=0
  output = [words[0]]
  while i < len(words) - 2:
    T_0 = words[i]
    T_1 = words[i+1]
    T_2 = words[i+2]
    key = (T_0, T_2)
    p_dist = prob[key]
    if len(p_dist) >1 and np.random.random() < 0.3:
      middle = sample_word(p_dist)
      output.append(T_1)
      # The original line was trying to append 3 elements, this has been modified to append a single formatted string
      output.append("<" + middle + ">")
      output.append(T_2)
      i += 2
    else:
      output.append(T_1)
      i += 1
  if i == len(words) - 2:
    output.append(words[-1])
  return detokenizer.detokenize(output)
```
